```python
# Step 1: Install and import libraries
!pip install pandas matplotlib seaborn --quiet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load CSV
df = pd.read_csv('/content/student_feedback.csv')
df.columns = df.columns.str.strip()

print("âœ… File loaded successfully!")
print("\nColumns found:\n", df.columns.tolist())
display(df.head())

# Step 3: Drop unnecessary column (like 'Unnamed: 0')
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Step 4: Select only numeric/rating columns
rating_cols = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]

# Step 5: Convert to numeric if not already
df[rating_cols] = df[rating_cols].apply(pd.to_numeric, errors='coerce')

# Step 6: Calculate averages
avg_ratings = df[rating_cols].mean().sort_values(ascending=False)

# Step 7: Visualize average ratings
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_ratings.values, y=avg_ratings.index, palette='coolwarm')
plt.title("Average Ratings per Feedback Category", fontsize=14)
plt.xlabel("Average Rating")
plt.ylabel("Feedback Category")
plt.xlim(0, 5)
plt.show()

# Step 8: Correlation heatmap (to see relationships)
plt.figure(figsize=(8, 6))
sns.heatmap(df[rating_cols].corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Correlation Between Feedback Metrics")
plt.show()

# Step 9: Suggest improvements automatically
print("\nðŸ’¡ Suggestions for Improvement:\n")
for col in avg_ratings.index:
    rating = avg_ratings[col]
    if rating < 3:
        print(f"ðŸ”´ '{col}' is rated low ({rating:.2f}). Suggest reviewing or improving this aspect.")
    elif 3 <= rating < 4:
        print(f"ðŸŸ  '{col}' is moderate ({rating:.2f}). Could use minor improvements.")
    else:
        print(f"ðŸŸ¢ '{col}' is performing well ({rating:.2f}). Keep it up!")

print("\nâœ… Visualization and suggestions complete.")
```
